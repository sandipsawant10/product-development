```python
y = df["Outcome"]
y
```

```python
x = df.drop(columns = "Outcome")
x
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size = 0.8, random_state = 10)
xtrain.size
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
model = lr.fit(xtrain, ytrain)
```

```python
ypred = lr.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(ytest, ypred)
acc
```

```python
from sklearn.metrics import classification_report
cr = classification_report(ypred, ytest)
print(cr)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(ypred, ytest)
cm
```

```python
import matplotlib.pyplot as plt
disp = ConfusionMatrixDisplay(confusion_matrix = cm)
disp.plot()
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
model2 = GNB.fit(xtrain, ytrain)
ypred2 = GNB.predict(xtest)
ypred2
```

```python
cr2 = classification_report(ypred2, ytest)
print(cr2)
```

```python
cm2 = confusion_matrix(ypred2, ytest)
cm2
```
